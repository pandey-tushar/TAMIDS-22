```
import numpy as np, pandas as pd, matplotlib.pyplot as plt
```

```
dept_df = pd.read_csv('primary_df_2.csv').iloc[:,1:]
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
dept_df['Departments'] = dept_df['Departments'].str[1:-1]

for i in range(dept_df.shape[0]):
  dept_df['Departments'][i] = dept_df['Departments'][i].replace("' ","'").replace("'","")

tmp_df = dept_df['Departments'].apply(lambda x: x.split(","))
array_out = mlb.fit_transform(tmp_df)
df_out = pd.DataFrame(data=array_out, columns=mlb.classes_)
df_out.columns = df_out.columns.str.replace("'","")
dept_df_2 = pd.concat([dept_df, df_out],axis =1).drop(['Departments'],axis =1 )
```

```
dept_df_2 = dept_df_2.drop(['Authors at my Institution','Research Organizations - standardized',
                         'Country of Research organization','Title','Title_abs'],axis=1)
```

```
dept_df_2['Times cited'] = dept_df_2['Times cited'].fillna(0)
```

```
dept_df_2
```

```
X = dept_df_2.iloc[:,2:]
X = X.iloc[:,(X.sum(axis=0)>2500).values]
X
```

```
import kmapper as km
from sklearn.manifold import TSNE
from sklearn.preprocessing import MinMaxScaler

mapper = km.KeplerMapper(verbose=1)


projected_X = mapper.fit_transform(X,
                                   projection = TSNE(), scaler = None)
from sklearn import cluster

```

```
graph = mapper.map(projected_X,
                   clusterer=cluster.DBSCAN(eps=0.2, min_samples=10, metric = 'cosine'),
                   cover=km.Cover(n_cubes=35, perc_overlap=0.3), remove_duplicate_nodes=True)
```

```
y = dept_df_2['Publication Date']
custom_t = 'Pub Date- '+ y.astype(str) + ', Dept- '+ dept_df.iloc[y.index]['Departments'].astype(str)
html = mapper.visualize(graph,
                        path_html="dept_collab_2.html",
                        title="dept_Collab",
                        X_names =  custom_t.values,
                        color_function_name = ['Dept'],
                        include_searchbar = True,
                        custom_tooltips = custom_t.values)
from kmapper import jupyter
jupyter.display("dept_collab_2.html")
km.draw_matplotlib(graph)
import matplotlib.pyplot as plt
plt.show()
```
